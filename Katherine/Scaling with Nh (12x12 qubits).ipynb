```python
%load_ext autoreload
%autoreload
import numpy as np
import tensorflow as tf
```

In this demo file, I calculate the learning criterion for different numbers of hidden units. The minimum learning criterion for the different numbers of hidden units is stored in the vector "lcminvec."

```python
from dset_helpers import load_essiqurke_data, create_tf_essiqurke_dataset
from train_data_2 import Train_w_Data
```

```python
lcminvec = []
```

```python
from train_data_2 import Train_w_Data

energy = []
variance = []
cost = []
learning_criterion_vec = []

def main():
    config = {
        'name': 'test', # Can name each experiment so that results are saved in a unique directory
        'seed': 1234,
        
        #System Parameters
        'L': 12,  # number of atoms
        'V': 7, 
        'Omega': 1,
        'delta': 1.00, # this varies
        
        #RNN Parameters
        'nh': 28,  # number of hidden units
        'lr': 1e-3,  # learning rate
        'trunc': 100, # related to interactions - keep at 100 for all-to-all interactions
        
        #Training Parameters
        'Data_epochs':1000, 
        'ns': 1000,
        'batch_size': 100, # divide dset into batches of this size
        'data_step': 100, # takes care of autocorrelations in QMC data
        
        'Print':True, 
        'Write_Data': False, # save results (helpful when going to larger systems)
        'Plot': True # plot training curves
        }
    
    return Train_w_Data(config,energy,variance,cost, learning_criterion_vec) # make sure this is the same function that is imported above


if __name__ == "__main__":
    model,e,v,c, lc, lcmin28 = main() # can also access e, v, c and make new plots after training is complete (ie to zoom in)
lcminvec.append(lcmin28)
```
